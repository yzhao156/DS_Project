```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta

cases = pd.read_csv("../../Data/2020_CA_Region_Mobility_Report.csv")
# cases = cases[(cases["sub_region_1"]=="Ontario")]
```

```python
Toronto = cases[cases["sub_region_2"]==("Toronto Division")]
Ottawa = cases[cases["sub_region_2"]==("Ottawa Division")]
frames = [Toronto, Ottawa]
Mobility_dimension = pd.concat(frames)
# Mobility_dimension[["sub_region_2"]].drop_duplicates()
Mobility_dimension["sub_region_2"] = Mobility_dimension["sub_region_2"].map(lambda x: x.split()[0])
Mobility_dimension.insert(0, "surrogate_key",  np.arange(len(Mobility_dimension)))
Mobility_dimension.head()
```

```python
Mobility_dimension.to_csv("Mobility_dimension.csv",index=False)
```
