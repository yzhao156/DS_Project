```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta

mobility = pd.read_csv("../../Data/2020_CA_Region_Mobility_Report.csv")
# cases = cases[(cases["sub_region_1"]=="Ontario")]
mobility = mobility.dropna(subset=["sub_region_2"])
```

```python
mobility["sub_region_2"].drop_duplicates()
```

```python
durhanMobility = mobility[(mobility["sub_region_2"].map(lambda x: "durham" in x.lower()))]

haltonMobility = mobility[(mobility["sub_region_2"].map(lambda x: "halton" in x.lower()))]
# durhanMobility = mobility[(mobility["sub_region_2"].map(lambda x: "durham" in x.lower()))]
peelMobility = mobility[(mobility["sub_region_2"].map(lambda x: "peel" in x.lower()))]
yorkMobility = mobility[(mobility["sub_region_2"].map(lambda x: "municipality of york" in x.lower()))]
ottawaMobility = mobility[(mobility["sub_region_2"].map(lambda x: "ottawa" in x.lower()))]
torontoMobility = mobility[(mobility["sub_region_2"].map(lambda x: "toronto" in x.lower()))]
frames = [durhanMobility, haltonMobility, peelMobility, yorkMobility, ottawaMobility, torontoMobility]
mobility = pd.concat(frames)
Mobility_dimension = mobility.fillna({'Outbreak_Related':'Unknown'})
```
