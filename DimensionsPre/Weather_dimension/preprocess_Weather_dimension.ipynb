```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta

dataPath = "../../Data/"
```

```python
Ottawa_Weather_2020 = pd.read_csv(dataPath+"DailyWeather/OttawaIntAirport2020.csv")
York_Weather_2020 = pd.read_csv(dataPath+"DailyWeather/NorthYork2020.csv")
Ottawa_Weather_2021 = pd.read_csv(dataPath+"DailyWeather/OttawaIntAirport2021.csv")
York_Weather_2021 = pd.read_csv(dataPath+"DailyWeather/NorthYork2021.csv")
# frames = [Ottawa_Weather, York_Weather, Peel_Weather]
frames = [Ottawa_Weather_2020, York_Weather_2020,Ottawa_Weather_2021,York_Weather_2021]
Weather_dimension = pd.concat(frames)
# Weather_dimension["Station Name"] = Weather_dimension["Station Name"].map(lambda x: "Ottawa" if (x=="OTTAWA INTL A") else "Toronto")
# Weather_dimension = Weather_dimension.drop(columns=['B', 'C'])
Weather_dimension = Weather_dimension[(Weather_dimension["Date/Time"] < "2021-03-01") & (Weather_dimension["Date/Time"] >= "2020-11-01")]
Weather_dimension = Weather_dimension.drop(columns=['Max Temp Flag', 'Min Temp Flag', 'Min Temp Flag', 'Mean Temp Flag', 'Heat Deg Days Flag', 'Data Quality', 'Cool Deg Days Flag', 'Total Precip Flag', 'Dir of Max Gust (10s deg)', 'Dir of Max Gust Flag', 'Spd of Max Gust (km/h)', 'Spd of Max Gust Flag'])
```

```python
Weather_dimension["Station Name"].drop_duplicates()
```

```python
precipArray = Weather_dimension["Total Precip (mm)"].values
for i in range(3,len(precipArray)):
    if (np.isnan(precipArray[i])):
        precipArray[i] = round(sum(precipArray[i-3:i])/3,1)
Weather_dimension["Total Precip (mm)"] = precipArray

maxArray = Weather_dimension["Max Temp (°C)"].values
for i in range(3,len(maxArray)):
    if (np.isnan(maxArray[i])):
        maxArray[i] = round(sum(maxArray[i-3:i])/3,1)
Weather_dimension["Max Temp (°C)"] = maxArray

minArray = Weather_dimension["Min Temp (°C)"].values
for i in range(3,len(minArray)):
    if (np.isnan(minArray[i])):
        minArray[i] = round(sum(minArray[i-3:i])/3,1)
Weather_dimension["Min Temp (°C)"] = minArray






tmpArray = Weather_dimension["Mean Temp (°C)"].values
for i in range(3,len(tmpArray)):
    if (np.isnan(tmpArray[i])):
        tmpArray[i] = round(sum(tmpArray[i-3:i])/3,1)
Weather_dimension["Mean Temp (°C)"] = tmpArray

tmpArray = Weather_dimension["Heat Deg Days (°C)"].values
for i in range(3,len(tmpArray)):
    if (np.isnan(tmpArray[i])):
        tmpArray[i] = round(sum(tmpArray[i-3:i])/3,1)
Weather_dimension["Heat Deg Days (°C)"] = tmpArray

tmpArray = Weather_dimension["Cool Deg Days (°C)"].values
for i in range(3,len(tmpArray)):
    if (np.isnan(tmpArray[i])):
        tmpArray[i] = round(sum(tmpArray[i-3:i])/3,1)
Weather_dimension["Cool Deg Days (°C)"] = tmpArray

tmpArray = Weather_dimension["Total Rain (mm)"].values
for i in range(3,len(tmpArray)):
    if (np.isnan(tmpArray[i])):
        tmpArray[i] = round(sum(tmpArray[i-3:i])/3,1)
Weather_dimension["Total Rain (mm)"] = tmpArray

tmpArray = Weather_dimension["Total Snow (cm)"].values
for i in range(3,len(tmpArray)):
    if (np.isnan(tmpArray[i])):
        tmpArray[i] = round(sum(tmpArray[i-3:i])/3,1)
Weather_dimension["Total Snow (cm)"] = tmpArray


tmpArray = Weather_dimension["Snow on Grnd (cm)"].values
tmpArray[0] = 0.
for i in range(4,len(tmpArray)):
    if (np.isnan(tmpArray[i])):
        tmpArray[i] = round(sum(tmpArray[i-3:i])/3,1)
Weather_dimension["Snow on Grnd (cm)"] = tmpArray



```
